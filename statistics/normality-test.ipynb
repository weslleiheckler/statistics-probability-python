### Importing libraries

```python
from scipy import stats
from scipy.stats import norm, skewnorm
import matplotlib.pyplot as plt
```

### Normality Test (Normal Distribution)
The normality test aims to identify whether the data are normally distributed.

```python
# creating a normal distribution with 1,000 elements
data = norm.rvs(size=1000)
```

#### Histogram

```python
plt.hist(data, bins=20)
plt.title('Distribution')
```

#### Normal Probability Diagram or Q-Q Plot

```python
fix, ax = plt.subplots()
stats.probplot(data, fit=True, plot=ax)
plt.show()
```

#### Shapiro-Wilk's Test

```python
# null hypothesis: the data is normally distributed
# in his case, the p-value is bigger than 0.05
# that is, there is no how to reject the null hypothesis
stats.shapiro(data)
```

### Normality Test (Non-Normal Distribution)

```python
data2 = skewnorm.rvs(4, size=1000)
```

#### Histogram

```python
plt.hist(data2, bins=20)
plt.title('Distribution')
```

#### Normal Probability Diagram or Q-Q Plot

```python
fix, ax = plt.subplots()
stats.probplot(data2, fit=True, plot=ax)
plt.show()
```

#### Teste de Shapiro-Wilk's Test

```python
# null hypothesis: the data is normally distributed
# in this case, the p-value is fewer than 0.05
# that is, the null hypothesis is rejected
stats.shapiro(data2)
```
